```python
# Exercise 2: String builder
# Goal: practice f-strings and string methods
# This simulates building an AWS resource summary — very close to real AI app work

service  = "EC2"
region   = "us-east-1"
count    = 7
instance_type = "t3.medium"
status   = "  running  "   # messy input with extra spaces

# Clean up the status string
clean_status = status.strip().upper()

# Build a formatted summary using f-strings
summary = f"Service: {service} | Region: {region} | Count: {count} | Status: {clean_status}| instance_type: {instance_type}"
print(summary)

# String checks
print(f"\nIs it in us-east? {'us-east' in region}")
print(f"Service starts with E: {service.startswith('E')}")
print(f"Summary length: {len(summary)} characters")

# Multiline string — this is exactly how you'll write AI prompts later
prompt_template = f"""
You are a cloud cost analyst.
Summarize the following:
- Service: {service}
- Region: {region}
- Running instances: {count}
- instance type: {instance_type}

Provide a one-line cost recommendation.
"""
print("\n--- This is what an AI prompt looks like ---")
print(prompt_template)

# CHALLENGE: add a variable for instance_type (e.g. "t3.medium")
# and include it in both the summary and the prompt template
```

```python
# Exercise 3: Lists (Simulating Document Chunks)
# Goal: Practice creating and accessing items in a list.
# In RAG, you take a big document and split it into smaller 'chunks' (a list of strings).

doc_chunks = [
    "RAG stands for Retrieval-Augmented Generation.",
    "It combines an information retrieval system with an LLM.",
    "Vector databases are used to store the embedded chunks."
]

# Print the entire list and its length
print(f"Total chunks: {len(doc_chunks)}")
print(doc_chunks)

# Accessing items (Remember: Python starts counting at 0)
print("\n--- Accessing Chunks ---")
print(f"First chunk: {doc_chunks[0]}")
print(f"Last chunk:  {doc_chunks[-1]}") # -1 is a handy trick to get the last item

# Adding a new chunk
doc_chunks.append("LangChain is a popular framework for building RAG.")
print(f"\nUpdated total chunks: {len(doc_chunks)}")

# TODO: Print the newly added chunk (the 4th item in the list)
# print(doc_chunks[???])
```

```python
# Exercise 4: Dictionaries (Simulating an LLM API Request)
# Goal: Practice Key-Value pairs.
# Dictionaries map 'keys' to 'values'. This is exactly how you configure LLMs.

llm_config = {
    "model": "gpt-4-turbo",
    "temperature": 0.7,
    "max_tokens": 500,
    "stream": True
}

print("--- LLM Configuration ---")
print(f"Using model: {llm_config['model']}")
print(f"Temperature (creativity): {llm_config['temperature']}")

# Changing a value
llm_config["temperature"] = 0.2  # Lower temperature for more factual, RAG-style answers
print(f"Updated Temperature: {llm_config['temperature']}")

# Adding a new key-value pair
llm_config["top_p"] = 0.9

# TODO: Add a new key called "api_key" to the dictionary with a fake string value (like "sk-12345")
# TODO: Print the entire dictionary to verify it was added.
```

```python
# Exercise 5: Putting it together (Simulating Chat History)
# Goal: A List containing Dictionaries.
# When you talk to an AI, you send the entire conversation history as a list of dictionaries.

chat_history = [
    {"role": "system", "content": "You are a helpful AI assistant specialized in Python."},
    {"role": "user", "content": "What is RAG?"},
    {"role": "assistant", "content": "RAG is a technique to ground LLMs on external data."}
]

print("--- Chat History ---")
# Loop through the history and print it nicely
for message in chat_history:
    print(f"{message['role'].upper()}: {message['content']}")

# TODO: Append a new dictionary to 'chat_history' simulating the user asking: "How do I use LangChain?"
# TODO: Print the updated chat_history using the same loop approach as above.
```
