```python
# Exercise 2: String builder
# Goal: practice f-strings and string methods
# This simulates building an AWS resource summary — very close to real AI app work

service  = "EC2"
region   = "us-east-1"
count    = 7
status   = "  running  "   # messy input with extra spaces

# Clean up the status string
clean_status = status.strip().upper()

# Build a formatted summary using f-strings
summary = f"Service: {service} | Region: {region} | Count: {count} | Status: {clean_status}"
print(summary)

# String checks
print(f"\nIs it in us-east? {'us-east' in region}")
print(f"Service starts with E: {service.startswith('E')}")
print(f"Summary length: {len(summary)} characters")

# Multiline string — this is exactly how you'll write AI prompts later
prompt_template = f"""
You are a cloud cost analyst.
Summarize the following:
- Service: {service}
- Region: {region}
- Running instances: {count}

Provide a one-line cost recommendation.
"""
print("\n--- This is what an AI prompt looks like ---")
print(prompt_template)

# CHALLENGE: add a variable for instance_type (e.g. "t3.medium")
# and include it in both the summary and the prompt template
```
